**1.Single Neuron, Single Feature**  

$$
y = w \cdot x + b
$$

---

**2.Single Neuron, `n` Features**

$$
y = w_1 x_1 + w_2 x_2 + \dots + w_n x_n + b = \sum_{i=1}^n w_i x_i + b
$$

Or in vector form (more compact):  
$$
y = \mathbf{w}^\top \mathbf{x} + b
\quad \text{where } \mathbf{w} = \begin{bmatrix} w_1 \\ w_2 \\ \vdots \\ w_n \end{bmatrix}, 
\ \mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}
$$

---

**3.`m` Neurons, `n` Features**

For neuron $j$ (where $j = 1, 2, \dots, m$):  
$$
y_j = w_{j1} x_1 + w_{j2} x_2 + \dots + w_{jn} x_n + b_j = \sum_{i=1}^n w_{ji} x_i + b_j
$$

In **matrix form**:  
$$
\mathbf{y} = W \mathbf{x} + \mathbf{b}
$$
where:
- $W \in \mathbb{R}^{m \times n}$ is the weight matrix (row $j$: $[w_{j1}, w_{j2}, \dots, w_{jn}]$)
- $\mathbf{x} \in \mathbb{R}^n$ is the input feature vector
- $\mathbf{b} \in \mathbb{R}^m$ is the bias vector
- $\mathbf{y} \in \mathbb{R}^m$ is the output vector

**Example (2 neurons, 3 features):**  
$$
\begin{bmatrix} y_1 \\ y_2 \end{bmatrix}
=
\begin{bmatrix}
w_{11} & w_{12} & w_{13} \\
w_{21} & w_{22} & w_{23}
\end{bmatrix}
\begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
+
\begin{bmatrix} b_1 \\ b_2 \end{bmatrix}
$$

````python
import torch
import torch.nn as nn
import torch.optim as optim

# For reproducible and consitant result
torch.manual_seed(42)
````

**Data preperation:**

````python
# Distances in miles for recent bike deliverires
distances = torch.tensor([
    [1.0],
    [2.0],
    [3.0],
    [4.0]
], dtype=torch.float32)

# Corresponding delivery times in minutes
times =torch.tensor([
    [6.96],
    [12.11],
    [16.77],
    [22.21]
], dtype=torch.float32)
````

**`nn.Linear:`**  

PyTorch’s `nn.Linear` expects a **2D tensor** of shape `(batch_size, in_features)` as input and outputs a **2D tensor** of shape `(batch_size, out_features)`.  

The **parameters of `nn.Linear` in PyTorch** are:

1. **`in_features`** – Number of input features.
2. **`out_features`** – Number of output features.
3. **`bias`** (optional, default `True`) – Whether to include a bias term.

These control the weight matrix of shape `(out_features, in_features)` and the bias vector of shape `(out_features,)` if used.  

The **weight matrix** of a PyTorch `nn.Linear` layer has shape:

$$
(\text{out\_features}, \text{in\_features})
$$


So if you have:

```python
layer = nn.Linear(in_features=10, out_features=5)
```

* `layer.weight.shape` → `(5, 10)`
* `layer.bias.shape` → `(5,)` (if `bias=True`)

It’s basically **one row per output feature**, each row containing weights for all input features.


**Model building:**

````python
model = nn.Sequential(nn.Linear(1,1))
````

````python
print(model[0].weight.shape)
print(model[0].bias.shape)
````

---
**`nn.Sequential` container:**   
- `nn.Sequential` is index-based by default.

You can access layers using integer indices:

```python
model = nn.Sequential(
    nn.Linear(1, 4),
    nn.ReLU(),
    nn.Linear(4, 1)
)

model[0]  # first layer (Linear)
model[1]  # ReLU
model[2]  # second Linear
```

You can also **name layers** if you want:

```python
model = nn.Sequential(
    ('fc1', nn.Linear(1, 4)),
    ('relu', nn.ReLU()),
    ('fc2', nn.Linear(4, 1))
)

model.fc1
model.fc2
```
---

**Training Model:**

````python
# Mean Squared Error (Calculate loss)
loss_function = nn.MSELoss()

# calculate gradients
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training loop
for epoch in range(500):
    # Reset the optimizer's gradients
    optimizer.zero_grad()

    # Make predictions (forward pass)
    outputs = model(distances)

    # Calculate the loss
    loss = loss_function(outputs, times)

    # Calculate gradinets (backward pass -> back propagation)
    loss.backward()

    # Update the model's parameters
    optimizer.step()

    # Print loss every 50 epochs
    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch + 1}: Loss = {loss.item()}")
````

* **Batch size**: The number of samples processed **together in one forward/backward pass**.
* **Number of samples per epoch**: The total number of samples in your dataset.

So:

$
\text{Number of batches per epoch} = \frac{\text{Total samples}}{\text{Batch size}}
$
